## Выбор метрики

### 1.1. посчитать среднее количество страниц, которые были просмотрены в рамках одной сессии

```python
def select_means_between_thresholds(x, down, up):
    """группировка по пользователям 
    с усреднением, с последующим исключением 
    по границам значений"""
    
    res = x.groupby(by='userID').mean()
    return res.loc[ (res['all_cl']>down) & (res['all_cl']<up), ['all_cl']].values
```

```python
def plot_two_hist(data1, data2, bins=50):
    _, ax = plt.subplots(figsize=(20,8))
    ax.hist(data1, bins=bins)
    ax.hist(data2, bins=bins, alpha=0.8)
```

```python
#считаем среднее кол-во просмотров по каждому пользователю,
#выбираем только тех пользователей которые осуществили переход по ссылке
#т.е. те пользователи у которых были просмотры

mean_exp_views = select_means_between_thresholds(exp, 0, outlier_threshold)
mean_control_views = select_means_between_thresholds(control, 0, outlier_threshold)

print('Выборочное Среднее эксперимет: {0:.3f}\
     \nВыборочное Среднее контроль:   {1:.3f}'
      .format(np.mean(mean_exp_views), np.mean(mean_control_views)))

plot_two_hist(mean_exp_views, mean_control_views, bins=30)
```

- распределения среднего кол-ва просмотров по пользователяем в тестовой и контрольной выборке 
имеет уже выраженную логнормальность
- отличия контроля и теста не видны

## Выбор стат. модели

#### проверка распределения и проведение статистического теста

```python
def fit_distribution(x, bins=10, distr_type='norm'):
    """подгонка теоретического распределения под эмпирическое"""
    
    dist = getattr(scipy.stats, distr_type)
    params = dist.fit(x)
    x_norm = np.linspace(min(x), max(x), bins)
    y = dist.pdf(x_norm, *params)
    
    return y, params

def fit_test_plot(x, bins, distr_type, ax, bin_edges, hist=None):
    """Подгонка распределения, К-С_тест, отрисовка графика"""
    
    y, params = fit_distribution(x, bins, distr_type=distr_type)
    #Kolmogorv-Smirnov test
    D, p_val = scipy.stats.ks_2samp(hist, y.flatten())
    RMSE = np.sum(np.power(hist - y.flatten(), 2.0)/len(bin_edges))
    label = 'p-level {0:.5f} D {1:.3f},  RMSE {2:.3f} of {3} distribution'.format(p_val, D, RMSE, distr_type)
    ax.plot(bin_edges[:-1], y, label=label)
    
    return y, params


def hist_of_true_theor(x, distr_type=None, bins=10,):
    """Вывод на график фактического и теоретических
    распределение с результатами теста"""
    
    hist, bin_edges = np.histogram(x, bins=bins, density=True)
    
    fig, ax = plt.subplots(1, 1, figsize=(20,8))
    ax.bar(bin_edges[:-1], hist, width=bin_edges[-1]/len(bin_edges[:-1]), alpha=0.3)
    
    y, params = None, None
    
    if not distr_type:
        distr_type = ['norm','lognorm','expon','weibull_min']

        for dist in tqdm_notebook(distr_type):
            fit_test_plot(x, bins, dist, ax, bin_edges, hist)
    else:
        y, params = fit_test_plot(x, bins, distr_type, ax, bin_edges, hist)
    
    plt.legend(loc='upper right', prop={'size': 14})
    
    return y, params
```

```python
BINS = 50
_, _ = hist_of_true_theor(mean_exp_views, bins=BINS)
```

```python
_, _ = hist_of_true_theor(mean_control_views, bins=BINS)
```

- видим, что в качестве статистической модели распределения <b>наиболее подходит Логнормальное распределение</b> 
- в качестве критерия согласованности используем непараметрический тест Колмогорова Смирнова, согласно которому теоретическое логнормальное распределение имеет самое меньшее расхождение с фактическим распределением из рассматриваемых распределений, к томуже <b>гипотеза о равенстве этих распределениий статистически значима и неотвергнута.</b>
- так же логнормальное теоретическое распределение <b>имеет наименьшую ошибку</b> по сравнению с другими распределениями
- это относиться как к тесту так и эксперименту

- воспользуемся свойством логнормального распределения: <b>"Если случайная величина имеет логнормальное распределение, то её логарифм имеет нормальное распределение"</b>

```python
exp_y, exp_params = hist_of_true_theor(np.log(mean_exp_views), bins=25, distr_type='norm')
```

```python
control_y, control_params = hist_of_true_theor(np.log(mean_control_views), bins=25, distr_type='norm')
```

- тесты снова показали значимый результат и верность Нулевой гипотезы о равенстве нормального теоретического и фактического распределений

### Выводы:  в качетсве стат модели будем использовать логарифм логнормального распределения который имеет нормальноее распределение, далее будем счиать метрики и брать моменты распределения уже по нему

### 2.Посмотреть распределения и особенности пользователей в эксперименте и на контроле.

- на этапе разведочного анализа мы уже увидели, что отношение просмотренных страниц к непросмотренным на даты  "2016-03-01" и "2016-02-29" отличается от остальных. Проверим это расчитав этот показатель.

#### Бутстреп оценки

```python
#расчет CTR
def ctr(series): return np.sum(series)/len(series)

def botstrap(series, metric, samples=1000, n=None, isProgres=True):
    """Бутстреп семплирование"""
    
    metrics = []
    if not n: n = len(series)//100
    iterable = range(samples) if isProgres else range(samples)
    for _ in iterable:
        sample_idx = np.random.randint(len(series), size=n)
        metrics.append(metric(series[sample_idx]))

    return metrics
```

```python
n_samples = 3000
#bootstrapping
ctr_botstrap_exp = botstrap(exp['all_cl'].values, ctr, samples=n_samples)
ctr_botstrap_control = botstrap(control['all_cl'].values, ctr, samples=n_samples)

#plot
plot_two_hist(ctr_botstrap_exp, ctr_botstrap_control, bins=30)
print("exp bootstrap mean {0:.4f}\ncontrol bootstrap mean {1:.4f}".format(
    np.mean(ctr_botstrap_exp), np.mean(ctr_botstrap_control)))

#test
mannwhitneyu = scipy.stats.mannwhitneyu(ctr_botstrap_exp, ctr_botstrap_control, alternative='greater')
print('means_diff {0:.4f},  Mann_Uhitney p-value:  {1:.6f}'.
      format(np.mean(ctr_botstrap_exp)-np.mean(ctr_botstrap_control), mannwhitneyu[1]))
```

### Выводы: Средние пользовательских переходов CTR значимо отличаются на контроле и в эксперименте

### 3.Посмотреть наличие различий по срезам, например, по браузеру. Если есть значимые различия, посчитать метрику - доля пользователей, которые не кликнули ни разу. Сделать выводы

```python
_, ax = plt.subplots(3,2,figsize=(20,15))
ax = ax.flatten()

for i,v in enumerate(tqdm_notebook(exp['browser'].unique())):
    test_botstrap = lambda data, select: botstrap(data.loc[data['browser']==select, :]['all_cl'].values, ctr, samples=10000)
    t1 = test_botstrap(exp, v)
    t2 = test_botstrap(control, v)
    mannwhitneyu = scipy.stats.mannwhitneyu(t1, t2, alternative='greater')
    test_text = 'Browser  {0}, Mann_Uhitney p-value:  {1:.6f}'.format(v, mannwhitneyu[1])
    ax[i].hist(t1, bins=50)
    ax[i].hist(t2, bins=50, alpha=0.8)
    ax[i].legend(('Experiment '+test_text, 'Control'),
           loc='upper center', shadow=True)
```

### Расчет CTR с группировкой по браузеру и времени

```python
ctr_grouped = lambda data_frame: data_frame.groupby(['ts','browser'])['all_cl'].agg(ctr).reset_index()

ctr_exp = ctr_grouped(exp)
ctr_control = ctr_grouped(control)
```

```python
def grouping_statistics(dataframe, columnm, values):
    """Группировка по пределенному признаку с построением бутстреп оценок
    dataframe - исходный датафрейм  
    columnm - колонка по которой группируются данные 
    values - значения колонки по которым делается группировка
    """
    
    statistics = []
    botstrap_samples = []
    for value in values:
        selected = dataframe.loc[dataframe[columnm]==value, 'all_cl'].values
        botstrap_sample = botstrap(selected, ctr, isProgres=False)
        statistics.append((value, len(botstrap_sample))+scipy.stats.norm.fit(botstrap_sample))
        botstrap_samples.append(botstrap_sample)
        
    return botstrap_samples, df(statistics, columns=(columnm,'size','mean','std'))

def compare_statistics(df_1, df_2, columnm, values, name_1, name_2):
    """Расчет бустреп оценок и проведение теста по каждой группе 
    в разрезе Контроль/Эксперимент"""
    
    b_1, df_1 = grouping_statistics(df_1, columnm, values)
    b_2, df_2 = grouping_statistics(df_2, columnm, values)
    
    stats = pd.concat([df_1.add_prefix(name_1+'_'), 
                       df_2[['mean', 'std']].add_prefix(name_2+'_')], axis=1)

    MU_test = lambda d1, d2: scipy.stats.mannwhitneyu(d1, d2, alternative='greater')
    stats = pd.concat([stats, 
                       df([np.round(MU_test(b_1[i], b_2[i]), decimals=4) for i in range(len(b_1))], 
                          columns=('MU_stat','p_val'))], 
                      axis=1)
    
    return stats
```

```python
compare_statistics(exp, control, 'browser', data['browser'].unique(), 'exp', 'control')
```

```python
compare_statistics(exp, control, 'ts', data['ts'].unique(), 'exp', 'control')
```
